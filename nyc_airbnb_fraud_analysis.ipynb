```python
borough_summary = (
    merged.groupby("neighbourhood_group")
    .agg(
        total_listings=("fraud_flag", "count"),
        total_fraudulent=("fraud_flag", "sum")
    )
    .reset_index()
)

print(borough_summary)
```

```python
borough_summary["fraud_rate_pct"] = (
    borough_summary["total_fraudulent"] / borough_summary["total_listings"] * 100)
```

```python
borough_summary = borough_summary.sort_values("fraud_rate_pct", ascending=False)

print(borough_summary)
```

```python
import scipy.stats as stats

contingency = pd.crosstab(merged['neighbourhood_group'], merged['fraud_flag'])

chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("p-value:", p)
```

```python
print(contingency)
contingency_norm = contingency.div(contingency.sum(axis=1), axis=0)
print(contingency_norm)
```

```python
# Levene's to test for variance across groups
from scipy.stats import levene

groups = [merged.loc[merged['neighbourhood_group']==g, 'price'] 
          for g in merged['neighbourhood_group'].unique()]
stat, p_val = levene(*groups, center='median')

print(f"Levene’s test statistic: {stat:.3f}, p-value: {p_val:.3f}")
```

```python
# Welch's ANOVA

import statsmodels.stats.oneway as oneway

groups = [merged.loc[merged['neighbourhood_group']==g, 'price'] 
          for g in merged['neighbourhood_group'].unique()]
welch = oneway.anova_oneway(groups, use_var='unequal')  
print(welch)
```

```python
# Logistic Regression
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# Features + target
X = merged[['price','minimum_nights','number_of_reviews','reviews_per_month',
            'availability_365','calculated_host_listings_count','beds',
            'neighbourhood_group','room_type']]
X = pd.get_dummies(X, columns=['neighbourhood_group','room_type'], drop_first=True)
y = merged['fraud_flag']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Baseline Logistic Regression
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)
print(confusion_matrix(y_test, clf.predict(X_test)))
print(classification_report(y_test, clf.predict(X_test)))

# Coefficients + odds ratios
coef_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': clf.coef_[0],
    'Odds_Ratio': np.exp(clf.coef_[0])
}).sort_values(by='Odds_Ratio', ascending=False)
print(coef_df)

# SMOTE resampling
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_train, y_train)

clf.fit(X_res, y_res)
print(confusion_matrix(y_test, clf.predict(X_test)))
print(classification_report(y_test, clf.predict(X_test)))
```
